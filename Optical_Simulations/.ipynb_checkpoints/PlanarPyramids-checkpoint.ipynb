```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import style
mpl.rcParams.update(mpl.rcParamsDefault)
import frgtools.misc as frgm
```

```python
fids = frgm.listdir()
```

```python
WL_0 = np.arange(300, 1501, 1)
WL_fill = np.zeros(1201)
```

```python
S = pd.read_csv('AM15.csv', delimiter=',', header=0)
WL_1, power_per_nm = S.iloc[:,0], S.iloc[:,1]
```

```python
T = pd.read_csv('testfile.txt', delim_whitespace=True, header=None)
WL_2, T_per_nm = T.iloc[:,0], T.iloc[:,1]
WL_2, T_per_nm = WL_2[::-1], T_per_nm[::-1]
```

```python
AM15 = np.interp(WL_0, WL_1, power_per_nm)
Trans = np.interp(WL_0, WL_2, T_per_nm)
for i in range(len(Trans)):
    if Trans[i] < 0:
        Trans[i] = 0
For_RT = Trans * AM15
```

```python
plt.figure(1)
plt.plot(WL_0, AM15, label='Top-cell')
RT = pd.DataFrame({'wl':WL_0, 'power': For_RT})

plt.plot(WL_0, For_RT, label='Bottom-cell')
plt.plot(WL_0, WL_fill, color='black')
plt.fill_between(WL_0, AM15, For_RT)
plt.fill_between(WL_0, For_RT, WL_fill)
RT = pd.DataFrame({'wl':WL_0, 'power': For_RT})



plt.legend()
plt.title('Incident Power Spectra ')
# plt.ylim(0,25)
# plt.xlim(0,1.2)

plt.ylabel('$W m^-2 * nm^-1$')
plt.xlabel('λ nm')
plt.show()
```

```python
with open('Import_to_RT.csv', 'a') as f:
    RT.to_csv(f, header=True)
```
