```python
import numpy as np
import os
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import style
mpl.rcParams.update(mpl.rcParamsDefault)
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
```

## Inspecting Material Data

```python
Gede = pd.read_csv('/Volumes/GoogleDrive/My Drive/Optical_Simulations/Material_data/IZO_n.csv', delimiter=',', header=None)
WL_n, n, WL_k, k = Gede.iloc[:,0], Gede.iloc[:,1], Gede.iloc[:,2], Gede.iloc[:,3]
```

```python
plt.plot(WL_n, n)
plt.plot(WL_k, k)
plt.show()
```

```python
df=[]
df=pd.DataFrame({'WL_n':[WL_n], 'n':[n], 'WL_k':[WL_k], 'k':[k],})
```

#### Smoothing Data

```python
from functools import partial #https://docs.python.org/2/library/functools.html
from scipy.signal import savgol_filter
window = 15
order = 11

#savgol_filter to smooth the data
df['n_smooth'] = df['n'].apply(
    partial(savgol_filter, window_length = window, polyorder = order)
)

df['k_smooth'] = df['k'].apply(
    partial(savgol_filter, window_length = window, polyorder = order)
)

k_smooth=df['k_smooth'][0]
n_smooth=df['n_smooth'][0]


plt.plot(np.array(WL_n),n_smooth, label='smooth_n')
plt.plot(WL_n, n, label='raw_n')

plt.plot(np.array(WL_k),k_smooth, label='smooth_k')
plt.plot(WL_k, k, label='raw_k')


plt.legend()
plt.show()
```

```python
WL_0 = np.linspace(310, 1200, 200)
k_adj = np.interp(WL_0, WL_k, k)
n_adj = np.interp(WL_0, WL_n, n)
```

```python
plt.plot(WL_0, n_adj)
plt.plot(WL_0, k_adj)
plt.show()
```

```python
IZO_export = pd.DataFrame({'wl':np.array(WL_0), 'n':n_adj, 'k':k_adj})
```

```python
# Gede_MAPI_Export
with open('IZO_v1.csv', 'w', newline='') as f:
    IZO_export.to_csv(f, header=True, index=False)
```
