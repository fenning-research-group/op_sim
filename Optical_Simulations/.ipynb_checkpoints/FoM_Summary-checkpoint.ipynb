## Creating a FoM summary

```python
FoM = pd.read_csv('FoM.csv', delimiter=',', header=0) # you have to manually create a csv with outputs
```

```python
FoM['Match']= FoM['Jsc_PSK'] - FoM['Jsc_Si']
FoM['Total_unlimited']= FoM['Jsc_PSK'] + FoM['Jsc_Si']
FoM['Total_limited']=np.nan
for n in range(0, len(FoM)):
    a = FoM['Jsc_PSK'][n]
    b = FoM['Jsc_Si'][n]
    if a<b:
        c = a
    if a>b:
        c = b
    if a == b:
        c= a
    FoM['Total_limited'][n] = c
result = FoM.pivot(index = 't', columns = 'Eg' , values='Total_limited')
```

```python
test = (FoM['Eg'], FoM['t'], FoM['Total_limited'])
ax = sns.heatmap(result, annot=True, fmt ='.2f', cmap="Greens", vmin=15, vmax=20)
plt.ylabel('Perovskite Thickness $(nm)$')
plt.xlabel('Perovskite Bandgap $(eV)$')
ax.invert_yaxis()
plt.title('Limited Tandem   $J_{sc}$ $ (mA)$')
plt.show()
```

```python
result2 = FoM.pivot(index = 't', columns = 'Eg' , values='Match')
ax = sns.heatmap(result2, annot=True, fmt ='.2f' ,cmap="RdBu", vmin=-10, vmax=10)
plt.ylabel('Perovskite Thickness $(nm)$')
plt.xlabel('Perovskite Bandgap $(eV)$')
ax.invert_yaxis()
plt.title('Tandem Current Mismatch Top-Bottom  $J_{sc}$ $ (mA)$')
plt.show()
```
